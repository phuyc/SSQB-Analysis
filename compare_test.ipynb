```python
import pandas as pd
```

```python
df = pd.read_csv('processed-reading-questions.csv')
df[df["reasoning_steps"] == -1].replace(-1, None)
df[df["distractor_complexity"] == -1].replace(-1, None)

# Now show summary statistics for each tests
practice_test_data = pd.read_csv('practice_test_data.csv')

# Filter out only the SAT questions
practice_test_data = practice_test_data[practice_test_data["Test"].str.contains(r"^SAT(1|2|3|4)")]

# Filter out the Reading & Writing section
practice_test_data = practice_test_data[practice_test_data["Subject"] == "Reading & Writing"]

# Exclude the Easy Module
practice_test_data = practice_test_data[practice_test_data["Module"] != 2]

# get questions in df that has questionId in practice_test_data
old_tests = df[df["questionId"].isin(practice_test_data["ID"])]

display(old_tests)
```

```python
new_tests = pd.read_csv('processed_new_test_questions.csv')
new_tests[new_tests["reasoning_steps"] == -1].replace(-1, None)
new_tests[new_tests["distractor_complexity"] == -1].replace(-1, None)
new_tests = new_tests.drop('Unnamed: 0', axis=1)
new_tests.head()
```

```python

combined_data = pd.concat([df, new_tests])
display(combined_data)
```
