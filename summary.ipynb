```python
import pandas as pd
```

```python
df = pd.read_csv('processed-reading-questions.csv')
df[df["reasoning_steps"] == -1].replace(-1, None)
df[df["distractor_complexity"] == -1].replace(-1, None)
df.head()
```

```python
# Distributions of correct answers
(df.groupby('correct_answer').size() / df['correct_answer'].count()) * 100
```

```python
# General statistics
df.describe().loc[['count', 'mean']].transpose()
```

```python
# Group by question type and calculate mean reasoning steps
mean_rs_critical_reading = df[df["skill_desc"].isin(['Command of Evidence', 'Inferences'])]['reasoning_steps'].mean()
mean_rs_others = df[~df["skill_desc"].isin(['Command of Evidence', 'Inferences'])]['reasoning_steps'].mean()

print("Mean reasoning steps for Command of Evidence and Inferences: ", mean_rs_critical_reading)
print("Mean reasoning steps for other question types: ", mean_rs_others)

# Display questions that require reasoning steps even though they are not Command of Evidence or Inferences 
display(df[~df["skill_desc"].isin(['Command of Evidence', 'Inferences']) & (df["reasoning_steps"] > 0)])
```

```python
# Now show summary statistics for each tests
practice_test_data = pd.read_csv('practice_test_data.csv')

# Filter out only the SAT questions
practice_test_data = practice_test_data[practice_test_data["Test"].str.contains(r"^SAT(1|2|3|4)")]

# Filter out the Reading & Writing section
practice_test_data = practice_test_data[practice_test_data["Subject"] == "Reading & Writing"]

# Exclude the Easy Module
practice_test_data = practice_test_data[practice_test_data["Module"] != 2]

# get questions in df that has questionId in practice_test_data
test_one_to_four = df[df["questionId"].isin(practice_test_data["ID"])]
```

```python
(test_one_to_four.groupby('correct_answer').size() / test_one_to_four['correct_answer'].count()) * 100
```

```python
test_one_to_four.describe().loc[['count', 'mean']].transpose()
```
